```python
txt_path = "/home/user15/RNN/dataset5/train.de"

length_list = []
with open(txt_path, 'r',encoding='utf-8') as f:
   
    for line in f:
        words = line.strip().split(" ")
        length_list.append(len(words))
        
```

```python
max(length_list)
```

```python
import matplotlib.pyplot as plt

# 샘플 데이터 (문장 길이 리스트)
# 히스토그램 그리기
plt.figure(figsize=(10, 6))
plt.hist(length_list, bins=10, edgecolor='black', alpha=0.75)
plt.title("Sentence Length Distribution", fontsize=16)
plt.xlabel("Sentence Length (Number of Words)", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
