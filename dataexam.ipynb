```python
def remove_empty_and_short_lines(file_path, output_path, min_length=26, min_words=6):
    """
    Removes empty lines, lines with fewer than `min_length` characters, 
    and lines with fewer than `min_words` words from the file.
    Saves the cleaned content to a new file.

    Args:
    file_path (str): Path to the input file to clean.
    output_path (str): Path to save the cleaned file.
    min_length (int): Minimum number of characters a line must have to be retained.
    min_words (int): Minimum number of words a line must have to be retained.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()

        # 필터링 조건 적용
        valid_lines = []
        for line in lines:
            stripped_line = line.strip()  # 공백 제거
            if stripped_line and len(stripped_line) >= min_length and len(stripped_line.split()) >= min_words:
                valid_lines.append(stripped_line + "\n")  # 줄바꿈 추가

        # 결과 파일 저장
        with open(output_path, 'w', encoding='utf-8') as output_file:
            output_file.writelines(valid_lines)

        print(f"Empty and short lines removed. Cleaned file saved to {output_path}.")
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except Exception as e:
        print(f"An error occurred: {e}")


# Example usage
file1 = "/home/user15/TT4/dataset9/training/newscom/training/news-commentary-v9.de-en.de"
file2 = "/home/user15/TT4/dataset9/training/newscom/training/news-commentary-v9.de-en.en"

file11 = "/home/user15/TT4/dataset0/news-commentary-v9.de-en.de"
file22 = "/home/user15/TT4/dataset0/news-commentary-v9.de-en.en"

remove_empty_and_short_lines(file1, file11, min_length=26, min_words=6)
remove_empty_and_short_lines(file2, file22, min_length=26, min_words=6)
```

```python
def check_empty_lines(file_path):
    """
    Checks if there are any empty lines in the file and prints the line numbers of empty lines.
    
    Args:
    file_path (str): Path to the file to check.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
            empty_line_indices = [i + 1 for i, line in enumerate(lines) if not line.strip()]
            
            if empty_line_indices:
                print(f"Empty lines found in {file_path}:")
                print(empty_line_indices)
            else:
                print(f"No empty lines found in {file_path}.")
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except Exception as e:
        print(f"An error occurred: {e}")


# Example usage:
file1 = "/home/user15/TT4/dataset0/news-commentary-v9.de-en.de"
file2 = "/home/user15/TT4/dataset0/news-commentary-v9.de-en.en"

check_empty_lines(file1)
check_empty_lines(file2)
```

```python
def filter_and_save(input_file_de, input_file_en, output_file_de, output_file_en):
    with open(input_file_de, 'r', encoding='utf-8') as f_de, \
        open(input_file_en, 'r', encoding='utf-8') as f_en, \
        open(output_file_de, 'w', encoding='utf-8') as out_de, \
        open(output_file_en, 'w', encoding='utf-8') as out_en:
        
        for line_de, line_en in zip(f_de, f_en):
            words_de = line_de.strip().split()
            words_en = line_en.strip().split()
            
            if len(words_de) >= 3 and len(words_en) >= 3:
                out_de.write(line_de.strip() + "\n")
                out_en.write(line_en.strip() + "\n")
                
input_file_de = '/home/user15/TT4/dataset0/News-Commentary.de-en.de'  # 독일어 데이터 파일 경로
input_file_en = '/home/user15/TT4/dataset0/News-Commentary.de-en.en'  # 영어 데이터 파일 경로
output_file_de = '/home/user15/TT4/dataset1/News-Commentary.de-en.de'   # 필터링된 독일어 파일
output_file_en = '/home/user15/TT4/dataset1/News-Commentary.de-en.en'    # 필터링된 영어 파일

filter_and_save(input_file_de, input_file_en, output_file_de, output_file_en)
```

```python
import re
import html

def clean_text(text):
    # 1. HTML 엔티티를 일반 문자로 변환
    text = html.unescape(text)

    # 2. 위키 문법 제거 (예: [[링크]] → 링크, [[링크|텍스트]] → 텍스트)
    text = re.sub(r'\[\[([^|\]]+)\|([^]]+)\]\]', r'\2', text)  # [[링크|텍스트]] 형식 처리
    text = re.sub(r'\[\[([^]]+)\]\]', r'\1', text)  # [[링크]] 형식 처리

    # 3. { { }} 제거 (중첩된 형식 포함)
    while re.search(r'\{\s*\{[^{}]+\}\s*\}', text):  # 중첩되지 않은 { { }} 패턴이 남아 있을 경우 반복적으로 제거
        text = re.sub(r'\{\s*\{\s*([^|{}]+)\s*\|\s*([^{}]+)\s*\}\s*\}', r'\2', text)  # { { meta | 내용 } } 같은 형식 처리
        text = re.sub(r'\{\s*\{\s*([^{}]+)\s*\}\s*\}', '', text)  # { { meta } } 같은 단순 { { 내용 } } 제거

    # 남아 있는 모든 { { }} 제거
    text = re.sub(r'\{\s*\{\s*[^}]+\s*\}\s*\}', '', text)

    text = re.sub(r"'\s+'", '"', text)

    # 4. 기타 불필요한 태그 제거 (예: 파일, 썸네일 관련 태그)
    text = re.sub(r'\[\[\s*Datei\s*:\s*[^]]+\]\]', '', text)  # 이미지 관련 태그 제거
    text = re.sub(r'\[\[\s*([^\]]+)\s*\]\]', r'\1', text)  # 남은 위키 링크 제거

    # 5. 불필요한 공백 정리 (개행 유지)
    text = re.sub(r'[ \t]+', ' ', text)  # 스페이스와 탭만 정리 (줄바꿈 유지)
    text = re.sub(r'^[ \t]+|[ \t]+$', '', text)  # 앞뒤 공백 제거 (줄바꿈 유지)

    return text

def process_files(input_file_de, input_file_en, output_file_de, output_file_en):
    with open(input_file_de, 'r', encoding='utf-8') as f_de, \
         open(input_file_en, 'r', encoding='utf-8') as f_en, \
         open(output_file_de, 'w', encoding='utf-8') as out_de, \
         open(output_file_en, 'w', encoding='utf-8') as out_en:

        for line_de, line_en in zip(f_de, f_en):
            cleaned_de = clean_text(line_de.rstrip('\n'))  # 개행 유지하면서 앞뒤 공백 제거
            cleaned_en = clean_text(line_en.rstrip('\n'))  # 개행 유지하면서 앞뒤 공백 제거

            out_de.write(cleaned_de + '\n')
            out_en.write(cleaned_en + '\n')

# 파일 경로 설정
input_file_de = '/home/user15/TT4/dataset2/train.de'
input_file_en = '/home/user15/TT4/dataset2/train.en'
output_file_de = '/home/user15/TT4/dataset3/train.de'
output_file_en = '/home/user15/TT4/dataset3/train.en'

# 파일 처리 실행
process_files(input_file_de, input_file_en, output_file_de, output_file_en)

print("파일 정리 완료: clean_train.de, clean_train.en")
```

```python
import re
import html

def clean_text(text):
    # 1. HTML 엔티티를 일반 문자로 변환
    text = html.unescape(text)

    # 2. 중첩된 { { ... }} 내부의 [[ ... ]] 패턴 정리 (반복 적용)
    while re.search(r'\{\s*\{[^{}]+\}\s*\}', text):  # 중첩된 { { ... }} 패턴이 남아 있으면 반복
        text = re.sub(r'\[\[\s*([^|\]]+)\s*\|\s*([^]]+)\s*\]\]', r'\2', text)  # [[A|B]] → B
        text = re.sub(r'\[\[\s*([^]]+)\s*\]\]', r'\1', text)  # [[A]] → A
        text = re.sub(r'\{\{\s*([^|{}]+)\s*\|\s*([^{}]+)\s*\}\}', r'\2', text)  # { { meta | 내용 } } → 내용
        text = re.sub(r'\{\{\s*([^{}]+)\s*\}\}', '', text)  # { { 내용 } } → 제거

    # 3. 남아 있는 모든 { { ... }} 제거
    text = re.sub(r'\{\{\s*[^}]+\s*\}\}', '', text)

    # 4. 위키 문법 정리 (반복적으로 `[[...]]` 패턴을 제거)
    while re.search(r'\[\[[^]]+\]\]', text):  # `[[...]]` 패턴이 남아 있을 경우 반복
        text = re.sub(r'\[\[\s*([^|\]]+)\s*\|\s*([^]]+)\s*\]\]', r'\2', text)  # [[USS Voyager | USS " Voyager "]] → USS " Voyager "
        text = re.sub(r'\[\[\s*([^]]+)\s*\]\]', r'\1', text)  # [[Volleyball]] → Volleyball

    # 5. 특정 파일 참조 제거 (Datei 등)
    text = re.sub(r'\[\[\s*Datei\s*:\s*[^]]+\]\]', '', text)  # [[Datei:파일명]] → 제거

    # 6. `[[...]]` 남은 패턴 완전 삭제
    text = re.sub(r'\[\[[^]]+\]\]', '', text)  # [[A]] → 삭제

    # 7. `|` 기호 제거 (이전 코드에서 누락됨)
    text = re.sub(r'\|', '', text)

    # 8. **대괄호 및 중괄호 제거 (공백 유지)**
    text = re.sub(r'[\[\]\{\}]', ' ', text)  # `[] { }` 제거 → **공백으로 대체하여 띄어쓰기 보존**

    # 9. ' ' → " " 로 변경
    text = re.sub(r"'\s+'", '"', text)

    # 10. 불필요한 공백 정리 (줄바꿈 유지)
    text = re.sub(r'\s+', ' ', text).strip()  # 여러 개의 공백을 하나로 줄임 (줄바꿈 유지)

    return text

def process_files(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f_in, \
         open(output_file, 'w', encoding='utf-8') as f_out:
        
        for line in f_in:
            cleaned_line = clean_text(line)
            f_out.write(cleaned_line + '\n')  # 줄바꿈 유지

# 파일 경로 설정
input_file1 = "/home/user15/TT4/dataset4/train.de"
output_file1 = "/home/user15/TT4/dataset5/train.de"

input_file2 = "/home/user15/TT4/dataset4/train.en"
output_file2 = "/home/user15/TT4/dataset5/train.en"

# 파일 처리 실행
process_files(input_file1, output_file1)
process_files(input_file2, output_file2)

print("변환 완료: train.de & train.en")
```

```python
def read_tail(file_path, n):
    """
    Reads the last n lines of a file.
    
    Args:
    file_path (str): Path to the file.
    n (int): Number of lines to read from the end of the file.
    
    Returns:
    list: A list of the last n lines.
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()  # Read all lines
        return lines[-n:] if len(lines) >= n else lines

output_valid1 = "/home/user15/TT4/dataset0/news-commentary-v9.de-en.de"
output_valid2 = "/home/user15/TT4/dataset0/news-commentary-v9.de-en.en"

a= read_tail(output_valid1, 1000)
b= read_tail(output_valid2, 1000)

for aa,bb in zip(a,b):
    print("#######")
    print(aa)
    print(bb)



```
